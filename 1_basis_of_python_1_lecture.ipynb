```python
# 外部データの読み込み

# pandas
import pandas as pd
data_dir = './sample_data/'
# csvを読み込む
data = pd.read_csv(data_dir+"some.csv")
# 好きな行を見る
print(data.iloc[0])
print(data.iloc[0:2])
# 好きな列を見る
print(data.iloc[:,0])
print(data.iloc[:,0:3])

print(data.columns)

# 項目を指定して見る
print(data.name)
print(data.height)

# 項目を指定して好きな行を見る
print(data.iloc[0].weight)
print(data.weight[1])

# 一行ずつ見る
for i in range(data.shape[0]): # shape は表の(行数,列数)、shape[0] は行数、shape[1] は列数を表す。
    print(data.iloc[i])

# 行を追加
s = pd.Series(['Jiro',30,75,180],index=data.columns)
data = data.append(s,ignore_index=True)

# 新規列名を指定して追加
data['BMI'] = [20, 30, 25]
print(data)

# 行、列を指定して編集
data.iloc[0,4] = 40
print(data)
```

```python
####################################
# 可視化
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
plt.style.use('ggplot')

temper_data=pd.read_csv(data_dir+"temperature_data.csv")

# データの中身をみる
print(temper_data.iloc[0])
print(temper_data.iloc[:3])
print(temper_data.iloc[:,0])
print(temper_data.iloc[:,1:3])

# 簡単な統計値を出す
print(temper_data.mean()) # 平均
print(temper_data.var()) # 分散
print(temper_data.std()) # 標準偏差

temper_data.describe() # 列ごとの統計量の比較
```

```python

#2つの文字を間に空白を入れてつなげる関数
def connectStr(str1,str2):
    a = str1 + ' ' + str2
    return( a )

a = connectStr('Hello','World!')
print( a )

# 2つのリストをつなげる関数
def connectList(list1, list2):
    a = list1 + list2
    return a

a = connectList([0,1,2],[1,2,3,4])
print( a )
```

```python
####################################
# 乱数
import random

#サイコロ
random.randint(1,6) # 1～6の整数をランダムに出力．　何が出るかはわからない！
random.randint(1,20) # 1～20の整数をランダムに出力．　何が出るかはわからない！

random.random() # 0 以上 1 未満の実数値を返す．何が出るかはわからない！
random.uniform(1.0, 10.0) # 1 以上 10以下の実数値を返す．何が出るかはわからない！

# break （くじびき）
import random
for i in range(5): # くじを5回までひけます
    number = random.randint(1, 10) # 1から10までの数字を1つもらえます
    if number == 1: # もし1が出たら・・・
        print( "atari !" )  # あたりです！
        break # 当たったので終わります
    else:
        print( "hazure" ) # はずれなのでもう一回引きましょう
```
