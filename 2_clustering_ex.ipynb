```python
#################################
# 演習１
import numpy as np
import pandas as pd
data_dir = './sample_data/'
f = pd.read_csv(data_dir+"height_weight.csv")
vlist = np.array(f)

# 類似度
def cossim(x,y):
    D = np.dot(x,y) # 内積
    xd = np.linalg.norm(x) #||x||
    yd = np.linalg.norm(y) #||y||
    s = D/(xd*yd) # コサイン類似度
    return s # 表示

# compute distances
A = np.array([180,75])
slist = []
for row in vlist:
    s = cossim(A,row[[2,3]])
    slist.append(s)

# visualize
import matplotlib.pyplot as plt
import matplotlib.cm as cm
sinds = np.argsort(slist)
count = 0
for i in sinds:
    row = vlist[i]
    c = cm.hot(count/len(vlist)) #色のリストを作成
    plt.scatter(row[2],row[3],color=c)
    count += 1
plt.scatter(A[0], A[1],c='green',marker='+',s=60)
```

```python
#################################
# 演習2
import pandas as pd
f = pd.read_csv(data_dir+'data.csv', header=None)
vlist = np.array(f) #計算のためにデータフレームを行列に変換

from sklearn.cluster import KMeans
#K = 2
km_model = KMeans(n_clusters=2) #K=4のkmeansモデルの初期化
km_model.fit(vlist) #実際にデータをクラスタリング

km_model.cluster_centers_ # クラスタの中心たち

labels=km_model.labels_ # 各データがどのクラスタかが入っている

plt.scatter(vlist[:,0], vlist[:,1], c=labels,cmap=cm.Accent)
plt.show()


#K = 3
km_model = KMeans(n_clusters=3) #K=4のkmeansモデルの初期化
km_model.fit(vlist) #実際にデータをクラスタリング

km_model.cluster_centers_ # クラスタの中心たち

labels=km_model.labels_ # 各データがどのクラスタかが入っている

plt.scatter(vlist[:,0], vlist[:,1], c=labels,cmap=cm.Accent)
plt.show()

#K = 4
km_model = KMeans(n_clusters=4) #K=4のkmeansモデルの初期化
km_model.fit(vlist) #実際にデータをクラスタリング

km_model.cluster_centers_ # クラスタの中心たち

labels=km_model.labels_ # 各データがどのクラスタかが入っている

plt.scatter(vlist[:,0], vlist[:,1], c=labels,cmap=cm.Accent)
plt.show()
```
