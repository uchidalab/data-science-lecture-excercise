```python
###############################
import pandas as pd
import matplotlib.pyplot as plt

# Read the iris dataset
iris = pd.read_csv(data_dir+'iris.csv')
# get the "petal width" data
X = iris['petal width']
plt.hist(X)
plt.show()

# substract the mean to all data to obtain the centered data
X_zero_mean = X-X.mean()
plt.hist(X_zero_mean)
plt.show()

# Divide by the standard deviation to normalize the data
X_scaled = X_zero_mean/X_zero_mean.std()
plt.hist(X_scaled)
plt.show()
```

```python
####################################
# 練習2
from sklearn.decomposition import PCA


X = iris.drop('class', axis=1)

# Normalize the data
X_scaled = (X-X.mean())/X.std()

# Run PCA
pca = PCA(n_components=4)
pca.fit(X_scaled)

# Plot the importance of each component
plt.plot(pca.explained_variance_ratio_)

print(pca.explained_variance_ratio_)
print(pca.components_)
```

```python
####################################
# 練習3

X = iris.drop('class', axis=1)

# Normalize the data
X_scaled = (X-X.mean())/X.std()

# Transform input data into an np.array
Y = np.array(X_scaled)

# Compute the covariance matrix
cov_matrix = np.cov(Y.transpose())

# Compute the eigen values and eigen vectors
(l,V) = np.linalg.eig(cov_matrix)

# you should obtain the same results as before, modulo the sign of the principal vectors
print("Explained variance ratio: ", l/l.sum())
print("the 4 principal vectors: ", V.transpose())

```
