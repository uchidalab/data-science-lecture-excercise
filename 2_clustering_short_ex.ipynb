```python
import numpy as np
####################
#練習１
x = np.array([3,5])
y = np.array([6,1])

#ユークリッド距離 (L2ノルム)
d = np.linalg.norm(x-y)
print(d)

#マンハッタン距離 (L1ノルム)
d1 = np.linalg.norm(x-y, ord=1)
print(d1)

#内積
ip = np.dot(x,y)
print(ip)

#コサイン類似度
s = np.dot(x,y)/(np.linalg.norm(x)*np.linalg.norm(y))
print(s)

x = np.array([3,5,2])
y = np.array([6,1,2])

#ユークリッド距離 (L2ノルム)
d = np.linalg.norm(x-y)
print(d)

#マンハッタン距離 (L1ノルム)
d1 = np.linalg.norm(x-y, ord=1)
print(d1)

#内積
ip = np.dot(x,y)
print(ip)

#コサイン類似度
s = np.dot(x,y)/(np.linalg.norm(x)*np.linalg.norm(y))
print(s)
```
