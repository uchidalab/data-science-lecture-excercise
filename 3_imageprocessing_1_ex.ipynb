```python
#### Exercice 1 ####
camera = data.camera() # load the image camera
io.imshow(camera)   # Display the image
io.show()

import numpy as np
nrows, ncols = camera.shape # Get the dimension of the image 
                            # (number of rows and number of columns)
row, col = np.ogrid[:nrows, :ncols] # Create a vectors of row indices 
                                    # and column indices
print( "Row indices: ", row )
print( "Column indices: ", col )
 
cnt_row, cnt_col = nrows / 2, ncols / 2 # Center of the disk
outer_disk_mask = ((row - cnt_row)**2 + (col - cnt_col)**2 > (nrows / 2)**2)
# outer_disk_mask has True values for pixel outside the disk 
# and False values for pixels inside the disk
camera[outer_disk_mask] = 0 # Put all pixel with True values 
                            # in outer_disk_mask to 0
io.imshow(camera)   # Display the image
io.show()
```

```python
# Exercice 3
from skimage import data
from skimage import io
import numpy as np
import matplotlib.pyplot as plt

image = data.astronaut().astype(float)
nrows, ncols, _ = image.shape

# 
result = np.zeros((int(nrows/2), int(ncols/2), 3))

# 
# 22->1(nrows/2 - 1)
for i in range(int(nrows/2 - 1)):
    for j in range(int(ncols/2 - 1)):
        #ch
        result[i,j,0] = (image[i*2,j*2,0] + image[(i+1)*2,j*2,0] + 
                    image[i*2,(j+1)*2,0] + image[(i+1)*2,(j+1)*2,0])/4.0
        result[i,j,1] = (image[i*2,j*2,1] + image[(i+1)*2,j*2,1] + 
                    image[i*2,(j+1)*2,1] + image[(i+1)*2,(j+1)*2,1])/4.0
        result[i,j,2] = (image[i*2,j*2,2] + image[(i+1)*2,j*2,2] + 
                    image[i*2,(j+1)*2,2] + image[(i+1)*2,(j+1)*2,2])/4.0
              
io.imshow(result.astype(np.uint8))
io.show()
```

```python
#####################################
####### Exercice 8 ##################
#####################################
import numpy as np
from skimage import data, io, exposure
from skimage.filters import sobel
from matplotlib import pyplot as plt

coins = data.coins()
plt.subplot(221)
io.imshow(coins)

# Contrast stretching
p2, p98 = np.percentile(coins, (2, 98))
img_rescale = exposure.rescale_intensity(coins, in_range=(p2, p98))

plt.subplot(222)
io.imshow(img_rescale)

# Equalization
img_eq = exposure.equalize_hist(coins)

plt.subplot(223)
io.imshow(img_eq)

# Adaptive Equalization
img_adapteq = exposure.equalize_adapthist(coins, clip_limit=0.03)

plt.subplot(224)
io.imshow(img_adapteq)
plt.show()
```

```python
#####################################
####### Exercice 9 ##################
#####################################
import numpy as np
from skimage import data, io
from skimage.filters import sobel
from skimage.color import label2rgb
from matplotlib import pyplot as plt
from scipy import ndimage as ndi

TEXT = io.imread(data_dir + 'Text.bmp')
segmented_img = np.zeros(TEXT.shape)
#100Truemask
mask = TEXT[:,:] > 100
#mask255
segmented_img[mask] = 255

fill_text = ndi.binary_fill_holes(segmented_img) #
label, count = ndi.label(fill_text) #
sizes = np.bincount(label.ravel())
#20mask
mask_sizes = sizes > 20
mask_sizes[0] = 0
text_cleaned = mask_sizes[label]

labeled_text, _ = ndi.label(text_cleaned)

#
image_label_overlay = label2rgb(labeled_text, image=TEXT)

io.imshow(image_label_overlay)
plt.show()
```

```python
#####################################
####### Exercice 10 ##################
#####################################
import numpy as np
from skimage import data, io
from skimage.filters import sobel
from skimage.color import label2rgb
from skimage.morphology import watershed
from matplotlib import pyplot as plt
from scipy import ndimage as ndi

TEXT = io.imread(data_dir + 'Text.bmp')
markers = np.zeros_like(TEXT)
#
markers[TEXT < 50] = 1 # background
markers[TEXT > 150] = 2 # foreground
elevation_map = sobel(TEXT) #edge
#watershedsegmentation
segmentation = watershed(elevation_map, markers)

#
segmentation = ndi.binary_fill_holes(segmentation - 1)
          
labeled_text, _ = ndi.label(segmentation)

#
image_label_overlay = label2rgb(labeled_text, image=TEXT)

io.imshow(image_label_overlay)
plt.show()
```
