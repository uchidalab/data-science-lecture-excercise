```python
# 練習1
from skimage import data
from skimage import io
data_dir = './data/'
coffee = io.imread(data_dir + "coffee.tiff")  # 画像を読み込み

io.imshow(coffee)  # 読み込んだ画像を表示

print( coffee.shape )  # 画像の形状 (高さ，横幅，チャンネル数: 色)
```

```python
# 練習2
cat = data.chelsea() # 猫の画像を読み込み
io.imshow(cat)  # 読み込んだ画像を表示
# 赤みがかった画素のみに変更を加えるためのマスクを作成
reddish = cat[:, :, 0] > 160 # redチャンネルの画素値が160より大きい場合True，それ以外はFalseとなる行列
# 赤みがかった画素をgreenに変更
cat[reddish] = [0, 255, 0]  # reddishがTrueの画素のみをgreen ([0, 255, 0]) に変更
io.imshow(cat)  # 変更結果を表示
```
