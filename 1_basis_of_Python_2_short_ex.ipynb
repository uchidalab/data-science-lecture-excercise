```python
data_dir = './Data/'
```

```python
####################
# 練習１
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
plt.style.use('ggplot')

temper_data=pd.read_csv(data_dir+"temperature_data.csv")

#他の地方の分布も見てみよう
#福岡、札幌、那覇それぞれのヒストグラムを作成
temper_data.plot(y='Fukuoka',kind='hist')
temper_data.plot(y='Sapporo',kind='hist')
temper_data.plot(y='Naha',kind='hist')

#bins と alpha を変えて３地点の分布を比較してみよう
temper_data.plot.hist(bins=10, alpha=0.5)
temper_data.plot.hist(bins=100, alpha=0.5)

#身長体重データでの統計値も見てみよう
HW_data=pd.read_csv(data_dir+"height_weight.csv", sep=",")
HW_data.mean()#平均
HW_data.var()#分散
HW_data.std()#標準偏差
temper_data.describe()
```
