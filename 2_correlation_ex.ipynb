```python
data_dir = './data/'
import numpy as np
import pandas as pd
import scipy.stats
import matplotlib.pyplot as plt
```

```python
## 演習1-2

from statsmodels import api as sm
iris=pd.read_csv(data_dir+'iris.csv')

# classごとにデータを分類
irisSetosa = iris[iris.iloc[:,4]=='Iris-setosa']
irisVersicolor = iris[iris.iloc[:,4]=='Iris-versicolor']
irisVirginica = iris[iris.iloc[:,4]=='Iris-virginica']

# class ごとに　sepal length と petal length  の回帰分析を行う
# setosa
plt.scatter(irisSetosa.iloc[:,0], irisSetosa.iloc[:,2])  # 分布を見る
X = irisSetosa['sepal length']  # 説明変数にsepal lengthを設定
X = sm.add_constant(X)  # 定数項を付ける
Y = irisSetosa['petal length']  # 目的変数にpetal lengthを設定
model = sm.OLS(Y,X)
result = model.fit()
print(result.summary())

plt.scatter(x=irisSetosa['sepal length'],y=irisSetosa['petal length']) and plt.plot(irisSetosa['sepal length'], 
model.fit().predict(X),c="red")

# versicolor
plt.scatter(irisVersicolor.iloc[:,0], irisVersicolor.iloc[:,2])  # 分布を見る
X = irisVersicolor['sepal length']
X = sm.add_constant(X)
Y = irisVersicolor['petal length']
model=sm.OLS(Y,X)
result = model.fit()
print(result.summary())

plt.scatter(x=irisVersicolor['sepal length'],y=irisVersicolor['petal length']) and plt.plot(irisVersicolor['sepal length'], 
model.fit().predict(X),c="red")

# virginica
plt.scatter(irisVirginica.iloc[:,0], irisVirginica.iloc[:,2])  # 分布を見る
X = irisVirginica['sepal length']
X = sm.add_constant(X)
Y = irisVirginica['petal length']
model=sm.OLS(Y,X)
result = model.fit()
result.summary()

plt.scatter(x=irisVirginica['sepal length'],y=irisVirginica['petal length']) and plt.plot(irisVirginica['sepal length'], 
model.fit().predict(X),c="red")
```

```python
######### 演習3 ###########
# 相関マップを作成し、最も相関が高い項目の組み合わせを見つけよ

iris = pd.read_csv(data_dir+'iris.csv')

# 相関行列: iris
R = np.corrcoef([iris.iloc[:,0], iris.iloc[:,1],
                 iris.iloc[:,2], iris.iloc[:,3]])
print(R)

# 最も相関が高い組み合わせをみつける
maxval = 0  # 相関の最大値を初期化
maxind = [-1,-1]  # 最も相関が高い組み合わせのindexのlistを初期化
for i in range(R.shape[0]):
    for j in range(R.shape[1]):
        if i == j:  # 対角成分の場合
            continue  # スキップ
        val = R[i,j]
        # valがmaxvalよりも大きい場合更新
        if val > maxval:
            maxval = val
            maxind = [i,j]
print(maxval)
print(maxind)
```

```python
######### 演習4 ###########
# P値マップ作成
P = []
for i in range(R.shape[0]):  # 行方向
    pp = []
    for j in range(R.shape[1]):  # 列方向
        # 各項目の組み合わせについてp値を算出
        _ ,p = scipy.stats.pearsonr(iris.iloc[:,i], iris.iloc[:,j])
        pp.append(p)
    P.append(pp)  # i行目のp値(pp)をPのi行目に挿入
        
# Pマップのヒートマップを描く
import seaborn as sns
sns.set(font_scale=1.5)  # フォントの大きさを調整
cr = ['SepalLength', 'SepalWidth', 
      'PetalLength','PetalWidth']  # 横軸と縦軸の意味
sns.heatmap(P, 
            annot=True,  # マップの実際の値を表示する
            annot_kws={'size': 15},  # 値のフォントの大きさ
            xticklabels=cr,  # 横軸の意味
            yticklabels=cr)  # 縦軸の意味
# グラフを表示する
plt.show()
```
